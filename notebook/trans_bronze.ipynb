```python
bronze_trans=spark.read.csv("/Volumes/fintrack/default/tables/transactions.csv",header=True,inferSchema=True)
bronze_trans.display()
```

```python
bronze_trans.write.format("delta") \
    .mode("overwrite") \
        .saveAsTable("fintrack.bronze.transaction")
```
