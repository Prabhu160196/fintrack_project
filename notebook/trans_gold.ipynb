1.Total revenue from successful payments

```python
Total revenue=spark.sql("select sum(amount)as Total_revenue from fintrack.silver.transaction where status='SUCCESS'")
Total revenue.display()
Total revenue.write.mode("overwrite").saveAsTable("fintrack_project_cat.gold.total_revenue")
```

2.Busiest and quietest city

```python
city=spark.sql('select city , count(txn_id)as transaction_count from fintrack.silver.transaction group by city order by transaction_count desc ')
city.display()
city.write.mode("overwrite").saveAsTable("fintrack.gold.Busiest_and_quietest_city")
```

3.Fraud count and percentage

```python
Fraud_count=spark.sql("select count(txn_id) as transaction_count, round(count(txn_id) /(select count( txn_id) from fintrack_project_cat.silver.transactions_silver) * 100,2) as percentage from fintrack.silver.transaction where is_fraud='true'")
Fraud_coun.display()
Fraud_coun.write.mode("overwrite").saveAsTable("fintrack.gold.fraudulent_transactions")
```

5.Row count difference between Version 0 and now

```python
Row_count=spark.sql("select count(*) as current_version_count ,(select count(*) from fintrack.silver.transaction version as of 1) as 0_version_count  from fintrack.silver.transaction ")
Row_count.display()
Row_count.write.mode("overwrite").saveAsTable("fintrack.gold.current_version_vs_0_version")
```

5.Average transaction amount by payment type

```python
Q5=spark.sql("select txn_type, round(avg(amount),2) as average_transaction_amount from fintrack.silver.transaction group by txn_type ")
Q5.display()
Q5.write.mode("overwrite").saveAsTable("fintrack.gold.average_transaction_amount")
```
